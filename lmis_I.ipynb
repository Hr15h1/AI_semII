```python
import pandas as pd
from tabulate import tabulate
```

Load and display the titanic dataset csv file


Here
PClass -> Booked ticket class
SibSp -> No of siblings or spouses aboard the titanic
Parch -> No of parents or children aboard the titanic


```python
df = pd.read_excel("datasets/titanic3.xls")
df_copy = df.copy()
print(tabulate(df, headers='keys', tablefmt="rounded_grid"))
```

Printing the rows which have non integer age values and less than 1(infants)

```python
nan_rows = df_copy[df_copy["age"].apply(lambda x: pd.notna(x) and isinstance(x, float) and not x.is_integer())]
print(tabulate(nan_rows, headers='keys', tablefmt="rounded_grid"))
```
